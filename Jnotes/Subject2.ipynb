```python
import pandas as pd
df_sub2_esense = pd.read_csv('..\sensor_data\eSense_sbj2.csv', names=['time','Device','1','2','3','4','5','6','Label'], index_col=0)
df_sub2_esense.index = pd.to_datetime(df_sub2_esense.index, unit='ms') + pd.Timedelta('02:00:00')
df_sub2_esense = df_sub2_esense.resample('10L', origin='start').first().interpolate()
df_sub2_esense.sort_index()
```

```python
df_sub2_ankle_dribbling = pd.read_csv('..\sensor_data\dribbling_sbj2_ankle.csv', index_col=0)
df_sub2_ankle_dribbling.index = pd.date_range(start='2021-04-14 19:26:42.914', periods = df_sub2_ankle_dribbling.shape[0], freq="10L")

df_sub2_ankle_layup = pd.read_csv('..\sensor_data\layup_sbj2_ankle.csv', index_col=0)
df_sub2_ankle_layup.index = pd.date_range(start='2021-04-14 19:16:25.794', periods = df_sub2_ankle_layup.shape[0], freq="10L")

df_sub2_ankle_shooting = pd.read_csv('..\sensor_data\shooting_sbj2_ankle_1.csv', index_col=0)
df_sub2_ankle_shooting.index = pd.date_range(start='2021-04-14 18:55:50.364', periods = df_sub2_ankle_shooting.shape[0], freq="10L")

df_sub2_ankle_movements = pd.read_csv('..\sensor_data\movements_sbj2_ankle_1.csv', index_col=0)
df_sub2_ankle_movements.index = pd.date_range(start='2021-04-14 19:36:19.244', periods = df_sub2_ankle_movements.shape[0], freq="10L")

df_sub2_ankle = pd.concat([df_sub2_ankle_dribbling, df_sub2_ankle_layup, df_sub2_ankle_shooting, df_sub2_ankle_movements])
#df_sub2_ankle.index = pd.date_range(start='2021-04-14 19:26:42.914000', periods = df_sub2_ankle.shape[0], freq="10L")
df_sub2_ankle.sort_index()
```

```python
df_sub2_wrist_dribbling = pd.read_csv('..\sensor_data\dribbling_sbj2_wrist.csv', index_col=0)
df_sub2_wrist_dribbling.index = pd.date_range(start='2021-04-14 19:26:24.464', periods = df_sub2_wrist_dribbling.shape[0], freq="10L")

df_sub2_wrist_layup = pd.read_csv('..\sensor_data\layup_sbj2_wrist.csv', index_col=0)
df_sub2_wrist_layup.index = pd.date_range(start='2021-04-14 19:16:05.604', periods = df_sub2_wrist_layup.shape[0], freq="10L")

df_sub2_wrist_shooting = pd.read_csv('..\sensor_data\shooting_sbj2_wrist.csv', index_col=0)
df_sub2_wrist_shooting.index = pd.date_range(start='2021-04-14 18:56:21.474', periods = df_sub2_wrist_shooting.shape[0], freq="10L")

df_sub2_wrist_movements = pd.read_csv('..\sensor_data\movements_sbj2_wrist.csv', index_col=0)
df_sub2_wrist_movements.index = pd.date_range(start='2021-04-14 19:36:45.074', periods = df_sub2_wrist_movements.shape[0], freq="10L")

df_sub2_wrist = pd.concat([df_sub2_wrist_dribbling, df_sub2_wrist_layup, df_sub2_wrist_shooting, df_sub2_wrist_movements])
#df_sub2_wrist.index = pd.date_range(start='2021-04-14 19:26:24.464000', periods = df_sub2_wrist.shape[0], freq="10L")
df_sub2_wrist.sort_index()
```

```python
df_sub2 = pd.concat([df_sub2_esense, df_sub2_ankle, df_sub2_wrist], axis=1, join="inner")
df_sub2
```
